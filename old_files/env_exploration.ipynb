Short notebook to explore and choose the environments

```python
import gymnasium as gym
import highway_env
import matplotlib.pyplot as plt
from IPython.display import clear_output

from config import config_dict
from config_continuous import config_dict as config_dict_continuous
from config_sdb3 import config_dict as config_dict_sdb3
```

1. Default env

```python
# Create and configure environment
env = gym.make("highway-fast-v0", render_mode="rgb_array")
env.unwrapped.configure(config_dict)  # Use unwrapped environment to configure
env.reset()

# Display the environment running
for _ in range(20):
    action = env.action_space.sample()
    observation, reward, terminated, truncated, info = env.step(action)
    clear_output(wait=True)
    plt.imshow(env.render())
    plt.show()
    if terminated:
        env.reset()
```
